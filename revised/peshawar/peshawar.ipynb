```python
import pandas as pd

# Load the data
file_path = 'peshawar.xlsx'
sheet_name = 'accident types'
df = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
import statsmodels.api as sm

# Define the dependent and independent variables
X = df[['Derailments', 'Collisions', 'Collisions at LC']]
y = df['No of Accidents']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the GLM model
glm_poisson = sm.GLM(y, X, family=sm.families.Poisson()).fit()

# Get the summary of the model
summary = glm_poisson.summary()
print(summary)
```

Collisions at Level Crossings:

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import glm

# Load the data
file_path = 'peshawar.xlsx'
sheet_name = 'Location of Accidents'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Replace spaces in column names with underscores
df.columns = df.columns.str.replace(' ', '_')


# Fit the GLM excluding the 'Year' variable
model = glm('No_of_Accidents ~ Accidents_at_Track + Accidents_in_Station_Limits',
            data=df,
            family=sm.families.Poisson()).fit()

# Print the summary of the model
print(model.summary())
```

## interpretation

Decreased by 8% in the number of total accidents. This result is nevertheless statistically significant which means that accidents at the track indeed are equally important when it comes to the total number of accidents.

The influence of accidents in Station limits
The coefficient of accidental frequency in station limits is 0. 2050. From this it can be inferred that a unit rise of the number of accidents within station limits leads to a rise of about 22. of them 8% in the total of the accidents. Summary
The findings of this model point to the fact that accidents that occur at the track greatly influence the number of accidents as compared to the influence of accidents that occur within the station limits.

# Type of Line

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import glm

# Load the data
file_path = 'peshawar.xlsx'
sheet_name = 'type of line'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Replace spaces in column names with underscores
df.columns = df.columns.str.replace(' ', '_')

# Reshape the data to a long format
df_long = pd.melt(df, id_vars=['Section'], value_vars=['Main_Line', 'Branch_Line'], 
                   var_name='Line_Type', value_name='No_of_Accidents')

# Convert 'Line_Type' to a categorical variable
df_long['Line_Type'] = df_long['Line_Type'].astype('category')

# Fit the GLM
model = glm('No_of_Accidents ~ Line_Type', 
            data=df_long,
            family=sm.families.Poisson()).fit()

# Print the summary of the model
print(model.summary())
```

## interpretation

The GLM analysis is focused on Main Line tracks’ ratio compared to Branch Line tracks regarding the number of accidents. The nominal variable line type is examined for its effect on the count of accidents using a Poisson distribution with a log link function.

Main Line
The coefficient for Line_Type[T. Main_Line] is 0. 5306. This means that, on the average, the Main Lines have the capability of terms of accident count compared to Branch Lines. More specifically, the annual rate of Main Lines Accidents is likely to hover at about forty-one. This is £3.80m or 1% lower than on Branch Lines. This is arrived at by taking the coefficient to the power of negative exponent as (e^-0. 5376=0. 589) indicating that Main Lines has a mean number of accidents that are about 59 percent of that on Branch Lines.

Branch Line
Branch Lines are the lowest level of categories in this model. Also the intercept value is equal to 2. 1401 is the logarithm of the anticipated accidents on Branch Lines. When thus exponentiated these translate to an expected number of approximately 8. 52 accidents on branch lines. This baseline estimate enables the participant Lines to be compared with Main Lines.


Summary
Main Lines are more favourable concerning the number of accidents that occur per year than Branch Lines with a decrease of about 41%. 1%. # Season

```python
import pandas as pd
import numpy as np  # Make sure to import numpy
from sklearn.linear_model import PoissonRegressor
from sklearn.preprocessing import StandardScaler

# Load and prepare the data
file_path = 'peshawar.xlsx'
sheet_name = 'Season'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Replace spaces in column names with underscores
df.columns = df.columns.str.replace(' ', '_')

# Add a small constant to avoid zero counts
df['No_of_Accidents'] = df['No_of_Accidents'] + 1

# Convert 'Season' to a categorical variable
df['Season'] = df['Season'].astype('category')

# Prepare data for scikit-learn
X = pd.get_dummies(df['Season'], drop_first=True)
y = df['No_of_Accidents']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit Poisson regression model
model = PoissonRegressor(alpha=1.0)
model.fit(X_scaled, y)

# Extract coefficients
coefficients = model.coef_
intercept = model.intercept_

# Print detailed summary
print("Poisson Regression Results")
print("=========================")
print(f"Intercept: {intercept:.4f}")

# Map the coefficients to feature names
feature_names = pd.get_dummies(df['Season'], drop_first=True).columns
for i, feature in enumerate(feature_names):
    print(f"Coefficient for {feature}: {coefficients[i]:.4f}")

# Print the exponentiated coefficients for interpretation
exponentiated_coefficients = np.exp(coefficients)
print("\nExponentiated Coefficients (Multiplicative Factors):")
for feature, coef in zip(feature_names, exponentiated_coefficients):
    print(f"{feature}: {coef:.4f}")

# Print the impact of each season in terms of multiplicative factors
baseline = "Winter"  # Assuming 'Winter' was dropped as baseline
print(f"\nComparing to Baseline Season ({baseline}):")
for feature, coef in zip(feature_names, exponentiated_coefficients):
    print(f"{feature}: {coef:.4f} times the number of accidents compared to {baseline}")
```

## interpretation 

We opted for Poisson regression instead of GLM due to practical issues such as perfect separation errors that can occur with GLM. Poisson regression in scikit-learn is straightforward for count data and avoids the complexities and potential fitting issues associated with GLMs in some cases. This approach simplifies the analysis while providing clear insights into the impact of different seasons on the number of accidents

The Intercept value is therefore 1. 9484 is the value of the quantity ‘accident rate for the baseline season on the log scale. This when exponentiated yields the base level of accidents for Winter season.

For the season-specific coefficients:

Next, the F statistic value for Spring is 0. 0481. Thus, log (number of accidents) indicating Spring is equal to 0. 1385 units higher than in Winter and within 0481 units of exceeding Summer. Raising this coefficient to the power of ex means predicting that the average number of accidents in Spring is approx 1. Thus, it has been found that the number of passengers during Spring is 0492 times that in Winter which shows a small rise.

Coefficient for Summer is zero. 2854. This implies that log number of Summer’s accidents is 0. 2854 units higher than in Winter season meaning that the higher the season the higher the probability of a network congestion. The resulting can be exponentiated to denote the fact that the estimated average of the number of accidents in Summer is, approximately 1. This constituted 3303 times that in Winter, a moderate rise.

The given figure of the season Winter being the starting point is equal to 1. It expanded 6079 times its size or to be better precise 6079 bigger than its initial size. This makes a confirmation that Winter is the reference category.

This implies that with each additional of the passenger train, there is an incremental change in the value of the log of the number of accidents of 0. 2226. This, when translated to a multiplicative factor means that each extra passenger train is accompanied by an increment of, roughly, twenty four. 9% increase in the expected number of accidents.PNG

Concerning the coefficient for Good, the result is equal to 0. 3094. This can be interpreted that for each additional goods train, the number of accidents as represented by the log number is raised by 0. 3094. Raising this figure to the power of goods train indicates that each additional goods train entails expenditure equivalent to about 36. Meaning an increase in the expected frequency of accidents by 2%. However, the coefficient of goods trains is insignificant implication that the effect of the goods trains on the number of accidents is less conclusive.

In general increasing the number of passenger trains shows that there is a strong relationship between the number of passenger trains and the number of accidents where for each increase in the number of these passenger trains the number of accidents increases notably. The impact of goods trains, however, cannot be easily determined mainly because the results are insignificant. By examining the model sum of squares value and comparing it to the residual sum of squares, one can say that the model bears a fair level of fit with data, which implies that the former accounts for a significant part of the variation in the numbers of the accidents.

# Severity

```python
import pandas as pd
import statsmodels.api as sm

# Load the data
file_path = 'peshawar.xlsx'
sheet_name = 'Severity'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Rename columns to remove extra spaces and make them consistent
df.columns = df.columns.str.strip().str.replace(' ', '_')

# Verify column names after renaming
# Define dependent variables and independent variable
X = df[['No_of_Accidents']]
y_casualties = df['No_Casulties']
y_deaths = df['No_of_Deaths']
y_injuries = df['No_of_Injuries']

# Add a constant to the independent variables matrix (for intercept)
X = sm.add_constant(X)

# Fit GLM model for each severity measure
model_casualties = sm.GLM(y_casualties, X, family=sm.families.Poisson()).fit()
model_deaths = sm.GLM(y_deaths, X, family=sm.families.Poisson()).fit()
model_injuries = sm.GLM(y_injuries, X, family=sm.families.Poisson()).fit()

# Print the summary for each model
print("Model Summary for Number of Casualties:")
print(model_casualties.summary())

print("\nModel Summary for Number of Deaths:")
print(model_deaths.summary())

print("\nModel Summary for Number of Injuries:")
print(model_injuries.summary())
```

## interpretation 
Number of Casualties Model

The GLM analysis for predicting the number of casualties reveals an intercept of 0.3582. This suggests that when there are no accidents, the expected number of casualties is roughly zero, although this is not statistically significant. The coefficient for the number of accidents is 0.1655, indicating that each additional accident is associated with approximately a 16.55% increase in the number of casualties. However, this relationship is not statistically significant, as indicated by the p-value of 0.169. The Pseudo R-squared value of 0.2723 suggests a moderate level of fit, meaning that the model explains a portion of the variability in the number of casualties based on the number of accidents.

Number of Deaths Model

For the number of deaths, the intercept is -1.0771, which is not statistically significant, with a p-value of 0.316. The coefficient for the number of accidents is 0.2944, suggesting a 29.44% increase in the number of deaths for each additional accident. This coefficient has a p-value of 0.109, which implies that while the effect is positive, it is not statistically significant at conventional levels. The Pseudo R-squared value of 0.3686 indicates that this model has a somewhat better fit compared to the casualties model, explaining more of the variability in the number of deaths based on the number of accidents.

Number of Injuries Model

In the model for predicting the number of injuries, the intercept is 1.2700, which is statistically significant with a p-value of 0.009, suggesting a baseline level of injuries. The coefficient for the number of accidents is 0.0733, indicating a smaller increase of about 7.33% in the number of injuries for each additional accident. This effect is not statistically significant, with a p-value of 0.442. The Pseudo R-squared value of 0.09346 shows that this model explains only a small proportion of the variance in the number of injuries.

Overall Interpretation

The combined analysis shows that while there is a positive association between the number of accidents and each type of severity (casualties, deaths, and injuries), the strength and significance of these relationships vary. The models indicate that the number of accidents does have some impact on casualties, deaths, and injuries, but the effects are not consistently strong or statistically significant. This suggests that additional factors or more complex interactions might be influencing these outcomes, and further exploration is needed to improve the models and understand the full scope of factors affecting railway accident severity.
